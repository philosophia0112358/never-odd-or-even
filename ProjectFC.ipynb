<h2><font color="green">SIZE</font></h2> *the goal is to convert the green col to a numeric type*

```python
data["Size"].unique() # NB here I prefer to use the [] notation instead of . notation to avoid confusion with .size attribute
```

*so we have 3 issues to address in order to clean the data in the Size column:*
    <ol>
    <li>*deal with the mysterious "+"*</li>
    <li>*replace "Varies with device" with some missing value (np.nan)*</li>
    <li>*transform all values ending with "M" (megabytes) and "k" (kilobytes) into the appropriate megabyte value*</li>
    </ol>

```python
data[data["Size"].str[-1] == "+"] # selecting the row/s where Size value ends with "+"
```

<p>*we obviously understand that since the Category value is missing all other values have been misplaced*</p>
<p>* in this case we simply decide to drop the row from the dataframe*</p>

```python
data.drop(10472, inplace=True)
```

```python
data["Size"] = data["Size"].replace("Varies with device", np.nan)
```

```python
data["Size"] = data["Size"].str.replace("M", "")
```

```python
data["Size"] = data["Size"].str.replace("k", "E-3")
```

```python
data["Size"] = data["Size"].astype(float)
```

*as we can see the dtype for col Size is float and the values are all expressed in Megabytes*

```python
data.dtypes
```

```python
data.head()
```

<h2><font color="green">INSTALLS</font></h2>
*the goal is to convert Installs col to a numeric dtype*

```python
data.Installs.unique()
```

```python
data.Installs.str[-1].unique()
```

*as we can see two actions must be taken:*
    <ol>
    <li>*Remove the "+" at the end of almost each value*</li>

<p>*now lets face the "Current Ver" and "Android Ver" columns*</p>
<p>*with the intention of converting both of them to a dotted number*</p>
<h2><font color="green">CURRENT VER & ANDROID VER</font></h2>

```python
data.columns = data.columns.str.replace(" ", "_") # so that I use dot notation efficiently (tabbing)
```

```python
data.Current_Ver.value_counts()
```

```python
data.Android_Ver.value_counts()
```

```python
data.Current_Ver = data.Current_Ver.replace("Varies with device", np.nan)
data.Android_Ver = data.Android_Ver.replace("Varies with device", np.nan)
```

```python
data[data.Current_Ver == "KM"] # since it makes no sense I will drop the row
```

```python
data.drop(7997, inplace=True)
```

```python
pattern = '^[0-9]\.[0-9]|[\d]|\W*'
temp1 = data.Current_Ver
temp1 = temp1.apply(lambda x : np.nan if str(x) == "nan" else re.findall(pattern, str(x))[0])
```

```python
data.Current_Ver = temp1
```

```python
data.Android_Ver = data.Android_Ver.str.replace(" and up", "")
data.Android_Ver = data.Android_Ver.str.strip("W")
```

```python
data
```

** lets create a df that maps apps to genres and one only for genres**

```python
data.Genres.unique()
```

```python
bridge = pd.DataFrame({"App":data.App, "Rating":data.Rating, "Installs":data.Installs, "Price":data.Price, "Genres":data.Genres})
```

```python
bridge.head() # I also imported Rating ... Price for future computations
```

```python
gen1 = bridge.Genres.apply(lambda x : x.split(";")[0])
gen2 = bridge.Genres.apply(lambda x : x.split(";")[-1])
bridge["Genres1"] = gen1
bridge["Genres2"] = gen2
```

```python
new_rows = bridge.query("Genres2 != Genres1") # 418 new rows
```

```python
new_rows.shape
```

```python
bridge = pd.concat([bridge, pd.DataFrame({"App":new_rows.App, "Rating":new_rows.Rating, "Installs":new_rows.Installs, "Price":new_rows.Price, "Genres1":new_rows.Genres2})], sort=True, ignore_index=True)
```

```python
bridge.shape # increased by 418
```

```python
bridge.head()
```

```python
bridge.drop(["Genres", "Genres2"], 1, inplace=True)
```

```python
bridge.rename(columns={"Genres1":"Genres"}, inplace=True)
```

```python
bridge.head()
```

```python
uni_gens = bridge.Genres.unique() 
```

*now lets add a col of type bool for each unique genres*

```python
i = 0
for gen in uni_gens:
    bridge[gen] = True if gen == bridge.Genres.values[i] else False
    i += 1
```
